```python
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document
```

```python
#step 1 your source documents
documents = [
    Document(page_content="Lnagchain helps developers build llm applications easily"),
    Document(page_content="Chroma is a vector database optimized for LLM-Based search"),
    Document(page_content="Embeddings covert text into high-dimentional vectors"),
    Document(page_content="Hugging face provides powerful embedding models")
]
```

```python
# step 2: Initialize embedding model
embedding_model = HuggingFaceEmbeddings(model="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# step 3: Create Chroma vector store in memory
vector_store = Chroma.from_documents(
    documents=documents,
    embedding=embedding_model,
    collection_name="my_collection"
)
```

```python
# step 4: Convert Vectorstore into retriever
retreiver = vector_store.as_retriever(search_kwargs={"k": 2})
```

```python
query="what is chroma used for"
results = retreiver.invoke(query)
```

```python
for i, doc in enumerate(results):
  print(f"\n---Result {i+1} ---")
  print(doc.page_content)
```
