```python
from langchain_community.retrievers import WikipediaRetriever
```

```python
# initialization of the retreiver (optional: set language and top_k)
retreiver = WikipediaRetriever(top_k_results=2, lang="en")
```

```python
#define the query
query = "Indian History"

#get relevant wikipedia document
docs = retreiver.invoke(query)
```

```python
docs
```

```python
#print retrived content
for i, doc in enumerate(docs):
  print(f"\n-----Result{i+1}---")
  print(f"content:\n{doc.page_content}...") # for truncate for display
```
